## Строение атома

### Первая наивная модель Томпсона
Было известно, что атомы излучают на строго определенных частотах. Из этого сделали вывод о том, что электрон
должен гармонически колебаться в атоме.

Если в атоме равномерно распределен положительных заряд, а в центре есть маленький электрон, то $F \sim r$.
Тогда колебания гармонические.

Эта модель точно не правильная, для доказательства рассмотрим опыт Резерфорда.
Из атома вылетает $\alpha$-частица. Направим поток $\alpha$ частиц на фольгу. Та состоит из электронов, с которыми
частицы почти не взаимодействуют. Но примерно 1 из 2000 частиц отскакивала от фольги, значит она врезалась в какую-то
тяжелую частицу. Значит протон такая частица.

На основании этого эксперимента была предложена
### Модель атома Бора
Если электрон летает около протона по круговой орбите, то он излучает. Если посчитать, это должно происходить за $\sim 10^{-8} c$.
Бор сказал:
1. есть основные энергетические уровни для электрона, на которых он не излучает
2. излучение происходит при переходи на другой уровень так, что $h\nu = E_{\text{нач}} - E_{\text{кон}}$

И при излучении и при поглащении света, энегрии квантуются.

Рассмотрим размерность пост. Планка (Дирака, которая с чертой).

$$ [\hslash] = \frac{\text{кг}\cdot\text{м}^2}{\text{с}} $$
Это же размерность момента импульса. Тогда предполагаем, что моменты импульса 
квантуются в атоме с таким шагом. На эксперименте это подтверждается.

$$mvr = n\hslash$$

$$r\frac{kq^2}{r} = \frac{mv^2}{r} = \frac{(n\hslash)^2}{mr^2}$$
$$r = \frac{n^2\hslash^2}{kq^2m}$$
Тогда энергия на $n$-ой орбите
$$ E = mv^2/2 - k\frac{q^2}{r} = -k\frac{q^2}{2r} = -\frac{k^2q^4m}{2\hslash^2}\cdot \frac{1}{n^2}$$
Коэффициент при $1/n^2$, $ \approx 13,6 \text{ эВ} $

Это энергия на первом уровне атома водорода.

Бальмер на основе экспериментальных данных вывел формулу для частот излучения атомами в видимом диапазоне.
$$ k(\frac{1}{2^2} - \frac{1}{n^2})$$

## Квантовые представления
В квантовой физике электрон тоже волна.
$$ E = \frac{hc}{\lambda}$$
$$\lambda = \frac{hc}{E}= \frac{h}{p}$$
Вводим эффективную длину волны для всех частиц по аналогии с фотоном. Это называется `длина волны де Бройля`.

Из этого условие стабильности электронной волны на орбите
$$ 2\pi r = n\lambda$$
Чтобы волна замыкалась на себя пройдя полный оборот.
